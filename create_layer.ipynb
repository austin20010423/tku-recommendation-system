```python
import tensorflow as tf
from keras import backend as K
import numpy as np
import pandas as pd

courses = pd.read_csv('/content/new_general_courses.csv')

report_arr = courses.values[:, 2].reshape(-1, 1)
report_arr = np.asarray(report_arr)

report_arr = K.cast_to_floatx(report_arr)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(1000, 64))
model.compile('rmsprop', 'mse')
out = model.predict(report_arr)

print(model.layers[0].get_weights())
```
